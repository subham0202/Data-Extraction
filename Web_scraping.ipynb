```python
import requests
import bs4
```

```python
resp=requests.get("https://www.mohfw.gov.in/")
```

```python
data=resp.text
```

```python
soup=bs4.BeautifulSoup(data,"html5lib")
```

```python
tag=soup.find("title")
print(tag)
print(tag.getText())
```

```python
tag=soup.find("marquee")
print(tag)
print(tag.getText())
```

```python
tag=soup.find("a")
print(tag)
print(tag.get("href"))
print(tag.getText())
```

```python
tag_list=soup.findAll("a")
print(len(tag_list))
```

```python
for tag in tag_list:
    print(tag.get("href"))
```

```python
tag_list=soup.findAll("strong",{'class':'mob-hide'})
for tag in tag_list:
    print(tag.getText())
```

```python
tag=soup.find("div",{'class':'fullbol'})
print(tag.getText())
```

```python
soup.findAll("span",{'class':['totalvac','coviddata']})
```

```python
tag_list=soup.findAll("img")
for tag in tag_list:
    print(tag)
```

```python
for tag in tag_list:
    print(tag.get("src"))
```

```python
img_urls=[]
for tag in tag_list:
    src=tag.get("src")
    if(src.startswith("https")):
        img_urls.append(src)
    else:
        img_urls.append("https://www.mohfw.gov.in/"+src)
print(img_urls)
```

```python
for url in img_urls:
    resp=requests.get(url)
    bytes_pixels=resp.content
    index=url.rindex("/")
    img_name=url[index+1:]
    file=open(f"images/{img_name}.png","wb")
    file.write(bytes_pixels)
    file.close()
```
